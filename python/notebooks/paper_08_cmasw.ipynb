# Paper 8: Pucci (2012) — CMASW Convexity

**Validates:** CC formula, grid across σ_asw × ρ, monotonicity, symmetry

```python
import sys; sys.path.insert(0, '..')
import math, numpy as np, matplotlib.pyplot as plt
from pricebook.viz import configure_theme
configure_theme(dark=False)

R_SWP = 0.0429; R_ASW = 0.0490; SIGMA_SWP = 0.30

def cc_ln(sigma_asw, rho, T=1.0, alpha_ratio=0.95):
    return R_ASW * (1 - alpha_ratio) * (math.exp(SIGMA_SWP * sigma_asw * rho * T) - 1)
```

```python
# CC heatmap
sigmas = [0.20, 0.30, 0.50]
rhos = [-0.90, -0.50, 0.0, 0.50, 0.90]

print(f"{'σ_asw':>8s}", ''.join(f'{r:>10.0%}' for r in rhos))
print('-' * 58)
for s in sigmas:
    row = [cc_ln(s, r) * 100 for r in rhos]
    print(f'{s:>8.0%}', ''.join(f'{v:>10.4f}' for v in row))

print('\nPaper range: -0.19% to +0.37%')
```

```python
# CC heatmap visualisation
fig, ax = plt.subplots(figsize=(10, 6))
    s_grid = np.linspace(0.05, 0.60, 40)
    r_grid = np.linspace(-0.95, 0.95, 40)
    S, R = np.meshgrid(s_grid, r_grid)
    CC = np.array([[cc_ln(s, r) * 10000 for s in s_grid] for r in r_grid])
    
    c = ax.contourf(S*100, R*100, CC, levels=20, cmap='RdBu_r')
    plt.colorbar(c, ax=ax, label='CC (bp)')
    ax.set_xlabel('ASW Vol (%)'); ax.set_ylabel('Correlation (%)')
    ax.set_title('CMASW Convexity Correction (bp)')
    ax.contour(S*100, R*100, CC, levels=[0], colors='black', linewidths=1)
    plt.tight_layout(); plt.savefig('paper_08_cmasw.png', dpi=150, bbox_inches='tight'); plt.show()
```
