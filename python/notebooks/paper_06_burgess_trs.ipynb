**Validates:** 3-component PV, coupon calculation, repo forward, recovery sensitivity

```python
import sys; sys.path.insert(0, '..')
import math, numpy as np, matplotlib.pyplot as plt
from pricebook.viz import configure_theme, apply_theme
configure_theme(dark=False)
```

```python
# Sample coupon
n_bonds = 80_736; face = 100; c = 0.0385
coupon = n_bonds * face * c * 0.5
print(f'Semi-annual coupon: ${coupon:,.2f}')
assert abs(coupon - 155_416.80) < 1
print('✓ Coupon matches paper')
```

```python
# Bond forward
B = 104.5433; r = 0.05
with apply_theme():
    fig, ax = plt.subplots(figsize=(10, 5))
    Ts = np.linspace(0, 2, 100)
    fwds = [B * (1 + r * t) for t in Ts]
    fwds_cont = [B * math.exp(r * t) for t in Ts]
    ax.plot(Ts, fwds, 'b-', linewidth=2, label='Simple: B(1+rT)')
    ax.plot(Ts, fwds_cont, 'r--', linewidth=2, label='Continuous: Be^(rT)')
    ax.set_xlabel('Time (years)'); ax.set_ylabel('Forward Price')
    ax.set_title('Bond Forward Price: Simple vs Continuous Repo'); ax.legend()
    plt.tight_layout(); plt.savefig('paper_06_trs.png', dpi=150, bbox_inches='tight'); plt.show()
```
