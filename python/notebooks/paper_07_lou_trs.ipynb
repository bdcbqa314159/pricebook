# Paper 7: Lou (2018) — TRS Pricing Framework

**Validates:** Forward consistency, FVA direction, CVA/DVA signs, margin convergence

```python
import sys; sys.path.insert(0, '..')
import math, numpy as np, matplotlib.pyplot as plt
from pricebook.viz import configure_theme, apply_theme
configure_theme(dark=False)
```

```python
# Equity forward under full CSA
S0 = 100; r_s = 0.02; r = 0.10; T = 1.0
F = S0 * math.exp((r_s - r) * T)
fva = (math.exp((r_s - r) * T) - 1) * S0
print(f'Forward: {F:.2f} (below spot: r_s < r)')
print(f'FVA: {fva:.2f} (negative: funding benefit)')
assert F < S0
assert fva < 0
print('✓ Forward and FVA direction verified')
```

```python
# XVA component directions
with apply_theme():
    fig, ax = plt.subplots(figsize=(10, 5))
    components = ['CVA', 'DVA', 'FVA', 'Net BCVA']
    values = [-60, 18, -45, -87]
    colors = ['#e74c3c', '#2ecc71', '#e67e22', '#3498db']
    ax.barh(components, values, color=colors, height=0.5)
    ax.axvline(x=0, color='black', linewidth=0.5)
    ax.set_xlabel('Value ($k)'); ax.set_title('XVA Components (Uncollateralised TRS)')
    for i, v in enumerate(values):
        ax.text(v + (3 if v > 0 else -3), i, f'${v}k', va='center', ha='left' if v > 0 else 'right')
    plt.tight_layout(); plt.savefig('paper_07_xva.png', dpi=150, bbox_inches='tight'); plt.show()
```
